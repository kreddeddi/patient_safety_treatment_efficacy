Read patients file - to create patients table

```python
# Read files using Autoloader with checkpoint
from pyspark.sql.functions import col

df_patients = (
    spark
    .readStream
    .format("cloudFiles")
    .option("cloudFiles.format", "csv")
    .option("pathGlobFilter", "*.csv")
    .option("header", "true")
    .option("cloudFiles.schemaLocation", "/Volumes/workspace/clinicaltrial_data/clinicaltrial_auto_landing/checkpoint/autoloader/1/")
    .load("/Volumes/workspace/clinicaltrial_data/clinicaltrial_auto_landing/autoloader_input/patients/")
)

```

Create db to save tables

```python
%sql
CREATE DATABASE IF NOT EXISTS clinicaltrial_bronze;
```

Write patients data to patients table

```python
# Write data to the delta table - clinicaltrial_bronze.patients

(
    df_patients
    .writeStream
    .format("delta")
    .option("checkpointLocation", "/Volumes/workspace/clinicaltrial_data/clinicaltrial_auto_landing/checkpoint/autoloader/1/")
    .trigger(availableNow=True)
    .toTable("clinicaltrial_bronze.patients")
)
```
